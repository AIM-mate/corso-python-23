## Domanda 2

Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date.

```python
def plot_stock(df, stock_name, start_date, end_date):
    df[stock_name].loc[start_date:end_date].plot()
    plt.legend([stock_name])
    plt.grid()


for colname in df.columns:
    plot_stock(df, colname, "2016-06-03", "2016-12-01")
    break
```

## Domanda 3

Converti il dataframe in NumPy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali nello stesso plot.

```python

df_np = df.iloc[:,:].to_numpy()
r = df_np[1:,:] / df_np[:-1,:] -1
r.astype(dtype = np.float64)

for ir in range(r.shape[1]):
    plt.plot(r[:, ir])
plt.xlabel('Time')
plt.ylabel('Dayling Return')
plt.grid()
```

## Domanda 4

Rispetto ai ritorni, calcola la media, deviazione standard, le correlazioni. 

Quale azione ha ritorni medi maggiori? Qual è la più volatile (maggiore deviazione standard)?

```python
r_mean = np.mean(r, axis=0)
print("Media azioni\n", r_mean)

r_std = np.std(r, axis=0, dtype=np.float64)
print("Deviazione standard azioni\n", r_std)

r_corr = np.corrcoef(r, rowvar=False)
```

```python
print(f"Ritorni medi maggiori: {df.columns[np.argmax(r_mean)]} con {np.round(np.max(r_mean)*100,3)} % al giorno")
print(f"Ritorni medi minori: {df.columns[np.argmin(r_mean)]} con {np.round(np.min(r_mean)*100,3)} % al giorno")
print(f"Maggiore volatilità: {df.columns[np.argmax(r_std)]} con {np.round(np.max(r_std),6)}")
print(f"Minore volatilità: {df.columns[np.argmin(r_std)]} con {np.round(np.min(r_std),6)}")
```

## Domanda 5

Un portafoglio è un insieme di assets, ad esempio:

* GOOGL 3 azioni
* AMZN 4 azioni
* FB 5 azioni

Il valore del portafoglio è 3 * prezzo di GOOGL + 4 * prezzo di AMZN + 5 * prezzo di FB

Costruisci il portafoglio che fa stonks nel 2020, ovvero che, se costruito il 01/01/2020 con 1000000€, ad oggi avrebbe il maggior valore, con le seguenti condizioni:

* all'inizio in ogni società non puoi investire più di 200000€ (20%)
* il numero di azioni acquistate di ogni società deve essere un intero non negativo
* non ci sono tassi di interesse, dividendi, commissioni e simili
* bisogna investire almeno il 99% della liquidità, la rimenente non verrà considerata

```python
# Esempio di portafoglio:
portfolio_test = pd.DataFrame(data={
    'code': ["GOOGL","AMZN","FB"],
    'num_stocks': [3,4,5],
    'value_2020_01_01': [1193.6639,1646.7944,182.9187],
    'value_2020_11_13': [1501.4839,2650.7724,234.636]
    })

print(portfolio_test)
```

```python
start_liquidity = 1e6
liquidity = start_liquidity
limit_per_stock = 0.2 * start_liquidity
mininvestment = 0.99 * start_liquidity

portfolio = pd.DataFrame(columns=['code', 'num_stocks', 'value_2020_01_01', 'value_2020_11_13'])

values_start = df.loc["2020-01-01"].to_numpy()
values_end = df.loc[df.index[-1]].to_numpy()

net_returns = values_end / values_start -1

for best_return in np.sort(net_returns)[-5:]:
    best_idx = np.where(net_returns == best_return)
    to_add = pd.DataFrame({
        'code': df.columns[best_idx][0],
        'num_stocks': np.floor(limit_per_stock / values_start[best_idx])[0],
        'value_2020_01_01': values_start[best_idx][0],
        'value_2020_11_13': values_end[best_idx][0]
        },
        index=[0])
    portfolio = pd.concat([portfolio, to_add], axis=0, ignore_index=True)

print(portfolio)
```

È già sopra il minimo?

```python
sum(portfolio["num_stocks"].to_numpy() * portfolio["value_2020_01_01"].to_numpy()) > mininvestment
```
